# To what extent did the relative frequency of nationalities change from modernism to postmodernism? To better answer this question, we compare the two csv files that we had created (results before 1980 and results on and after 1980). 

```python
import pandas as pd

f1 = pd.read_csv('C:/Users/alexa/OneDrive/Documenten/Collecting Data & Tools and Methods/Subquestion2before1980.csv')
f2 = pd.read_csv('C:/Users/alexa/OneDrive/Documenten/Collecting Data & Tools and Methods/Subquestion2onafter1980.csv')

df1 = (f2[f2.nationality.isin(f1.nationality)])
df1 = df1.sort_values(["percentage"], ascending=[0])
df2 = (f1[f1.nationality.isin(f2.nationality)])
```

# Note: an important criteria for this comparison was that we could only select nationalities that were present in both the first subset as well as the second subset. Although there was significant overlap between the two subsets (both subsets for example had 'american' as their most frequent nationality), it is possible that some nationalities that were present after 1980 but not before 1980, or vice versa, are excluded from this comparison.

# However, as a quick run through the countries absent in in either subset 1 or subset 2 shows, the most significant nationalities (with the highest counts and percentages) are still included in the comparison.

# The most significant exception to this rule is 'Ivorian', with 482 artworks from on and after 1980 but none before 1980. Other interesting nationalities that had artworks from on and after 1980, but not before, are 'native american', 'palestinian', kenyan' and 'albanian'. 

```python
list1 = list(df2.nationality)
list2 = list(f2.nationality)
left_out_1 = []

for item in list2:
    if item not in list1:
        left_out_1.append(item)

for item in left_out_1:
    print(f2.loc[f2["nationality"] == item])
```

# Some nationalities that were represented only by artworks from before 1980, but not on and after 1980, are 'latvian', 'costa rican', 'burkinabe' and 'canadian inuit'. 

```python
list3 = list(df1.nationality)
list4 = list(f1.nationality)
left_out_2 = []

for item in list4:
    if item not in list3:
        left_out_2.append(item)

for item in left_out_2:
    print(f1.loc[f1["nationality"] == item])
```

# Therefore, we here continue with the comparison of subset 1 and subset 2 (comparing only those countries which were present in either subset)

```python
output2 = list(df1['count'])
output3 = list(df1.percentage)

dictionary1 = {'nationality_onafter' : output2, 'percentage_onafter' : output3}

output4 = list(df2.nationality)
output5 = list(df2['count'])
output6 = list(df2.percentage)

dictionary1 = {'count_onafter' : output2, 'percentage_onafter' : output3}

dictionary2 = {'nationality' : output4, 'count_before' : output5, 'percentage_before' : output6}

df1 = pd.DataFrame(dictionary1)

df2 = pd.DataFrame(dictionary2)
```

```python
join_df = df2.join(df1)
join_df
```

# We also add columns 'count_difference', which calculates the difference in counts of nationality from before 1970 to on and after 1970, and 'percentage_difference', which calculates the percentage increase or decrease in nationality as a percentage of the total from before 1970 to on and after 1970. 

```python
join_df['count_difference'] = (join_df['count_onafter'] - join_df['count_before'])
join_df
```

```python
join_df['percentage_difference'] = (join_df['percentage_onafter'] - join_df['percentage_before']) / join_df['percentage_before'] * 100
join_df
```

```python
join_df = join_df.sort_values(["count_before"], ascending=[0])
```
